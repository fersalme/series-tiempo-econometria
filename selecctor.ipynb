## Selector de estudiantes

```python
import numpy as np

estudiantes = ["KEYNER", "MARIA", "MATEO", "VALENTINA", "YARIMA",
                "ALEXANDER", "AMPARO", "FABIANA", "NESTOR", "WENDY", "MISAEL"]
estudiante_sustentar = []
estudiante_preguntar = []

def escoger_sustentar():
    disponibles = [e for e in estudiantes if e not in estudiante_sustentar] 
    if len(disponibles) == 0:
        return "Ninguno, todos los estudiantes ya sustentaron."
    r_estudiante = np.random.choice(disponibles)
    estudiante_sustentar.append(r_estudiante)
    return r_estudiante

def escoger_preguntar():
    disponibles = [e for e in estudiantes if e not in estudiante_preguntar]
    if len(disponibles) == 0:
        return "Ninguno, todos los estudiantes ya preguntaron."
    r_estudiante = np.random.choice(disponibles)
    estudiante_preguntar.append(r_estudiante)
    return r_estudiante
```

```python
print(f"Estudiante a sustentar :  {escoger_sustentar()}")
```

```python
print(f"Estudiante pregunta sustentación : {escoger_preguntar()}")
```
